# Load data

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
train_data = pd.read_csv("/kaggle/input/int303-a2/train.csv")
test_data = pd.read_csv("/kaggle/input/int303-a2/test.csv")
```

# Data preprocessing

```python
# Delete columns with only one value 
train_data.drop(columns=['EmployeeCount'])
test_data.drop(columns=['EmployeeCount'])
train_data.drop(columns=['Over18'])
test_data.drop(columns=['Over18'])
train_data.drop(columns=['StandardHours'])
test_data.drop(columns=['StandardHours'])
```

```python
# Remove rows containing missing values
train_data_df = pd.DataFrame(train_data).dropna()
if 'Attrition' in test_data.columns:
    test_data = test_data.drop(columns=['Attrition'])
test_data_df = pd.DataFrame(test_data).dropna()

# Remove duplicate rows
train_data_df = train_data_df.drop_duplicates()
test_data_df = test_data_df.drop_duplicates()
# print(test_data_df)
```

```python
# One Hot
if 'Gender' in train_data_df.columns and 'Gender' in test_data_df.columns:
    Gender_train = train_data_df["Gender"]
    Gender_test = test_data_df["Gender"]

    Gender_train_array = pd.DataFrame(Gender_train).values
    Gender_test_array = pd.DataFrame(Gender_test).values

    # Create the OneHotEncoder object and fit the training data:
    encoder = OneHotEncoder(sparse=False)
    # One Hot
    encoded_Gender_train = encoder.fit_transform(Gender_train_array)
    encoded_Gender_test = encoder.fit_transform(Gender_test_array)

    # Convert encoded data into a list of columns
    encoded_list_Gender_train = list(encoded_Gender_train.flatten())
    encoded_list_Gender_test = list(encoded_Gender_test.flatten())
    encoded_list_Gender_train = [int(x) if x != float('inf') else x for x in encoded_list_Gender_train]
    encoded_list_Gender_test = [int(x) if x != float('inf') else x for x in encoded_list_Gender_test]

    train_data_df["Encoded Gender"] = pd.Series(encoded_list_Gender_train)
    train_data_df.drop("Gender", axis=1, inplace=True)

    test_data_df["Encoded Gender"] = pd.Series(encoded_list_Gender_test)
    test_data_df.drop("Gender", axis=1, inplace=True)
# OverTime
if 'OverTime' in train_data_df.columns and 'OverTime' in test_data_df.columns:
    mapping = {
        "Yes": 1,
        "No": 0
    }

    # 使用map()方法将"OverTime"列的值转换为对应的编码值
    train_data_df["Encoded OverTime"] = train_data_df["OverTime"].map(mapping)
    test_data_df["Encoded OverTime"] = test_data_df["OverTime"].map(mapping)

    train_data_df.drop("OverTime", axis=1, inplace=True)
    test_data_df.drop("OverTime", axis=1, inplace=True)
# JobRole
if 'JobRole' in train_data_df.columns and 'JobRole' in test_data_df.columns:
    JobRole_train = train_data_df["JobRole"]
    JobRole_test = test_data_df["JobRole"]

    JobRole_train_array = pd.DataFrame(JobRole_train).values
    JobRole_test_array = pd.DataFrame(JobRole_test).values

    encoder_2 = OneHotEncoder(sparse=False)
    encoded_JobRole_train = encoder_2.fit_transform(JobRole_train_array)
    encoded_JobRole_test = encoder_2.fit_transform(JobRole_test_array)

    encoded_list_JobRole_train = list(encoded_JobRole_train.flatten())
    encoded_list_JobRole_test = list(encoded_JobRole_test.flatten())
    encoded_list_JobRole_train = [int(x) if x != float('inf') else x for x in encoded_list_JobRole_train]
    encoded_list_JobRole_test = [int(x) if x != float('inf') else x for x in encoded_list_JobRole_test]

    train_data_df["Encoded JobRole"] = pd.Series(encoded_list_JobRole_train)
    train_data_df.drop("JobRole", axis=1, inplace=True)

    test_data_df["Encoded JobRole"] = pd.Series(encoded_list_JobRole_test)
    test_data_df.drop("JobRole", axis=1, inplace=True)
# EducationField
if 'EducationField' in train_data_df.columns and 'EducationField' in test_data_df.columns:
    EducationField_train = train_data_df["EducationField"]
    EducationField_test = test_data_df["EducationField"]

    EducationField_train_array = pd.DataFrame(EducationField_train).values
    EducationField_test_array = pd.DataFrame(EducationField_test).values

    encoder_3 = OneHotEncoder(sparse=False)
    encoded_EducationField_train = encoder_3.fit_transform(EducationField_train_array)
    encoded_EducationField_test = encoder_3.fit_transform(EducationField_test_array)

    encoded_list_EducationField_train = list(encoded_EducationField_train.flatten())
    encoded_list_EducationField_test = list(encoded_EducationField_test.flatten())
    encoded_list_EducationField_train = [int(x) if x != float('inf') else x for x in encoded_list_EducationField_train]
    encoded_list_EducationField_test = [int(x) if x != float('inf') else x for x in encoded_list_EducationField_test]

    train_data_df["Encoded EducationField"] = pd.Series(encoded_list_EducationField_train)
    train_data_df.drop("EducationField", axis=1, inplace=True)

    test_data_df["Encoded EducationField"] = pd.Series(encoded_list_EducationField_test)
    test_data_df.drop("EducationField", axis=1, inplace=True)
# Department
if 'Department' in train_data_df.columns and 'Department' in test_data_df.columns:
    Department_train = train_data_df["Department"]
    Department_test = test_data_df["Department"]

    Department_train_array = pd.DataFrame(Department_train).values
    Department_test_array = pd.DataFrame(Department_test).values

    encoder_4 = OneHotEncoder(sparse=False)
    encoded_Department_train = encoder_4.fit_transform(Department_train_array)
    encoded_Department_test = encoder_4.fit_transform(Department_test_array)

    encoded_list_Department_train = list(encoded_Department_train.flatten())
    encoded_list_Department_test = list(encoded_Department_test.flatten())
    encoded_list_Department_train = [int(x) if x != float('inf') else x for x in encoded_list_Department_train]
    encoded_list_Department_test = [int(x) if x != float('inf') else x for x in encoded_list_Department_test]

    train_data_df["Encoded Department"] = pd.Series(encoded_list_Department_train)
    train_data_df.drop("Department", axis=1, inplace=True)

    test_data_df["Encoded Department"] = pd.Series(encoded_list_Department_test)
    test_data_df.drop("Department", axis=1, inplace=True)
# BusinessTravel
if 'BusinessTravel' in train_data_df.columns and 'BusinessTravel' in test_data_df.columns:
    BusinessTravel_train = train_data_df["BusinessTravel"]
    BusinessTravel_test = test_data_df["BusinessTravel"]

    BusinessTravel_train_array = pd.DataFrame(BusinessTravel_train).values
    BusinessTravel_test_array = pd.DataFrame(BusinessTravel_test).values

    encoder_5 = OneHotEncoder(sparse=False)
    encoded_BusinessTravel_train = encoder_5.fit_transform(BusinessTravel_train_array)
    encoded_BusinessTravel_test = encoder_5.fit_transform(BusinessTravel_test_array)

    encoded_list_BusinessTravel_train = list(encoded_BusinessTravel_train.flatten())
    encoded_list_BusinessTravel_test = list(encoded_BusinessTravel_test.flatten())
    encoded_list_BusinessTravel_train = [int(x) if x != float('inf') else x for x in encoded_list_BusinessTravel_train]
    encoded_list_BusinessTravel_test = [int(x) if x != float('inf') else x for x in encoded_list_BusinessTravel_test]

    train_data_df["Encoded BusinessTravel"] = pd.Series(encoded_list_BusinessTravel_train)
    train_data_df.drop("BusinessTravel", axis=1, inplace=True)

    test_data_df["Encoded BusinessTravel"] = pd.Series(encoded_list_BusinessTravel_test)
    test_data_df.drop("BusinessTravel", axis=1, inplace=True)
# MaritalStatus
if 'MaritalStatus' in train_data_df.columns and 'MaritalStatus' in test_data_df.columns:
    MaritalStatus_train = train_data_df["MaritalStatus"]
    MaritalStatus_test = test_data_df["MaritalStatus"]

    MaritalStatus_train_array = pd.DataFrame(MaritalStatus_train).values
    MaritalStatus_test_array = pd.DataFrame(MaritalStatus_test).values

    encoder_6 = OneHotEncoder(sparse=False)
    encoded_MaritalStatus_train = encoder_6.fit_transform(MaritalStatus_train_array)
    encoded_MaritalStatus_test = encoder_6.fit_transform(MaritalStatus_test_array)

    encoded_list_MaritalStatus_train = list(encoded_MaritalStatus_train.flatten())
    encoded_list_MaritalStatus_test = list(encoded_MaritalStatus_test.flatten())
    encoded_list_MaritalStatus_train = [int(x) if x != float('inf') else x for x in encoded_list_MaritalStatus_train]
    encoded_list_MaritalStatus_test = [int(x) if x != float('inf') else x for x in encoded_list_MaritalStatus_test]

    train_data_df["Encoded MaritalStatus"] = pd.Series(encoded_list_MaritalStatus_train)
    train_data_df.drop("MaritalStatus", axis=1, inplace=True)

    test_data_df["Encoded MaritalStatus"] = pd.Series(encoded_list_MaritalStatus_test)
    test_data_df.drop("MaritalStatus", axis=1, inplace=True)
# Over18
if 'Over18' in train_data_df.columns and 'Over18' in test_data_df.columns:
    mapping = {
        "Y": 1,
        "N": 0
    }

    # Use the map() method to convert the values in the "OverTime" column to the corresponding encoded values.
    train_data_df["Encoded Over18"] = train_data_df["Over18"].map(mapping)
    test_data_df["Encoded Over18"] = test_data_df["Over18"].map(mapping)

    train_data_df.drop("Over18", axis=1, inplace=True)
    test_data_df.drop("Over18", axis=1, inplace=True)
```
